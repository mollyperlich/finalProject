```python
# Dependencies
import numpy as np
import pandas as pd
import requests
import time
```

```python
#get list of company names
company_names = pd.read_csv('companylist2.csv')
```

```python
#get symbols of the companies
companies = list(company_names['Symbol'])
```

```python
#only run once!
list_stocks = []
```

```python
num = 423
total = len(companies)
for i in range(402,total):
    time.sleep(14)
    url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol="+ f'{companies[i]}'+"&outputsize=full&apikey=WBX7LU6ERVSP822O"
    stock_response = requests.get(url)
    print(stock_response)
    stock_response = stock_response.json()
    meta_data = stock_response['Meta Data']
    del stock_response['Meta Data']
    list_stocks.append(stock_response)
    print(i)
```

```python
#create data frame for first list item and add to it in for loop
stock_data = pd.DataFrame(list_stocks[0]).iloc[-730:]

#rename column with company name
stock_data = stock_data.rename(columns={"Time Series (Daily)": f"{companies[0]}"})

for i in range(1, 575):
    #only get past 2 years
    stock_data[f'{companies[i]}'] = pd.DataFrame(list_stocks[i]).iloc[-730:]
```
