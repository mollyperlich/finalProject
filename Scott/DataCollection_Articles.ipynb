```python
sourceData = ""

with open('articles_Jan1_2019_Feb28_2019.json') as jsonFile:
    sourceData = json.load(jsonFile)

    
```

```python

response = requests.request("GET", sourceUrl)


# Check Response
if (response.status_code == requests.codes.ok):
    
    print(response)
```
